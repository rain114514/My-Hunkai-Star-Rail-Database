```python
file1 = '1.千面英雄.txt'
file2 = '2.手机输入.txt'
with open(file2, 'r', encoding = 'utf-8') as f: lines = f.readlines()
with open(file1, 'a', encoding = 'utf-8') as f:
    for line in lines: # 每行为"Num.Boss1/Boss2/Boss3"
        line = line.strip().split('.')
        bosses = line[1].split('/')
        f.write(' '.join(bosses) + '\n')
```
