```python
file0 = '0.货币战争.txt'
file1 = '1.手机输入.txt'
with open(file1, 'r', encoding = 'utf-8') as f: lines = f.readlines()
with open(file0, 'a', encoding = 'utf-8') as f:
    for line in lines: # 每行为"Num.Tag1/Tag2/Tag3/Tag4"
        line = line.strip().split('.')
        tags = line[1].split('/')
        f.write(' '.join(tags) + '\n')
```
