```python
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

file0 = '0.零和博弈.txt'
```

```python
def get_tag_to_count(): # 构建词缀-数量对应表
    with open(file0, 'r', encoding = 'utf-8') as f: lines = f.readlines()
    tag_to_count = {}
    for line in lines: # 每行为"tag1 tag2 tag3 tag4"
        tags = line.strip().split()
        for tag in tags: tag_to_count[tag] = tag_to_count.get(tag, 0) + 1
    return tag_to_count

tag_to_count = get_tag_to_count()
```

```python
def draw(tag_to_count): # 绘制饼状图
    tag_to_count = sorted(tag_to_count.items(), key = lambda x: x[1], reverse = True)
    tag = [item[0] for item in tag_to_count]; count = [item[1] for item in tag_to_count]
    plt.figure(figsize = (10, 10)); plt.pie(count, labels = tag, autopct = '%1.1f%%')
    plt.show('货币战争-零和博弈: 负面词缀分布'); plt.show()

draw(tag_to_count)
```
